### Set coordinate list

```python
coordinate_list = []
```

## Target site 

```python
browser.get("https://gps-coordinates.org/")
```

## Locate button

```python
buttons = browser.find_elements_by_tag_name("button")
target_button = buttons[0]
```

## Locate coord boxes

```python
latitude_box = browser.find_element_by_id("latitude")
longitude_box = browser.find_element_by_id("longitude")
address_box = browser.find_element_by_id("address")
```

## Import Keys

```python
from selenium.webdriver.common.keys import Keys
```

## Clear before initial run

```python
address_box.clear()
latitude_box.clear()
longitude_box.clear()

```

```python
import numpy
```

```python
numpy.random.uniform(.8,1.2)
```

## Test Send

```python
# Wait 5 second before starting
time.sleep(5)

import datetime

start = datetime.datetime.now()
count  = 0
for address in address_list:
    
    time.sleep(numpy.random.uniform(.8,1.2))
    
    address_box.send_keys(address)
    
    time.sleep(numpy.random.uniform(.8,1.2))
    
    target_button.click()
    
    time.sleep(numpy.random.uniform(.94,1.2))
    
    latitude = latitude_box.get_attribute("value")
    longitude = longitude_box.get_attribute("value")
    
    coordinate_list.append((longitude, latitude))
    
    address_box.clear()
    latitude_box.clear()
    longitude_box.clear()
    
    count = count + 1
    print("{} Address {} found coordinates {} {}".format(count, address, longitude, latitude))
    
print("Finished processing",len(address_list),"addresses in",datetime.datetime.now()-start)
```
